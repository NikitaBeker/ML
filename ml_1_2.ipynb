```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
x = pd.read_csv("https://raw.githubusercontent.com/koroteevmv/ML_course/main/ML1.2_regression/data/0_x.csv",
                header=None)
y = pd.read_csv("https://raw.githubusercontent.com/koroteevmv/ML_course/main/ML1.2_regression/data/0_y.csv",
                header=None)

x.head(),y.head()
```

```python
binary = LinearRegression()
binary.fit(x[[1]], y)
plt.scatter(x[1], y)
xx = np.linspace(x[1].min(), x[1].max(), 100).reshape((-1, 1))
plt.plot(xx, binary.predict(xx), c='r')

binary.coef_,binary.score(x[[1]],y)
```

```python
multiple = LinearRegression()
multiple.fit(x, y)
multiple.score(x, y)
```

```python
yy = multiple.predict(x)
plt.scatter(yy, y)
plt.plot(yy, yy, c='r')
```

```python
#   

poly = PolynomialFeatures(degree=2)


X_poly = poly.fit_transform(x[[1]])

model_poly = LinearRegression().fit(X_poly, y)

print(model_poly.coef_)
```

```python
#      
r2_poly = model_poly.score(X_poly, y)
print(r2_poly)

r2>r2_poly
```
