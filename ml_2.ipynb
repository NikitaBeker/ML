```python
import pandas as pd

df = pd.read_excel('/content/Вариант 2.xlsx')
#1,2
print(df.describe())
```

```python
#3
below_avg = len([k for k in df.iloc[2:-1,9] if k<df.iloc[-1,9]])
percent_below_avg = below_avg / (len(df)-2) * 100
print(f'Процент учащихся, выполнивших работу ниже среднего: {percent_below_avg:.2f}%')
```

```python
#4
failed_num = len([k for k in df.iloc[2:-1,9] if k<27])
percent_failed = failed_num / (len(df)-2) * 100
print(f'Процент учащихся, не сдавших экзамен: {percent_failed:.2f}%')
```

```python
#5

import matplotlib.pyplot as plt

passed = df.iloc[2:-1,9][df.iloc[2:-1,9] >= 27].count()
failed = df.iloc[2:-1,9][df.iloc[2:-1,9] < 27].count()

plt.pie([passed, failed], labels=['Сдавшие', 'Не сдавшие'])
plt.show()
```

```python
#6
df.iloc[2:-1,9].plot(kind='density')
plt.show()
```

```python
#7
A_grade_percent=len([k for k in df.iloc[2:-1,9] if k>=85])/len(df.iloc[2:-1,9])*100
B_grade_percent=len([k for k in df.iloc[2:-1,9] if (k>=70 and k<85)])/len(df.iloc[2:-1,9])*100
C_grade_percent=len([k for k in df.iloc[2:-1,9] if (k>=50 and k<70)])/len(df.iloc[2:-1,9])*100
D_grade_percent=len([k for k in df.iloc[2:-1,9] if (k<50)])/len(df.iloc[2:-1,9])*100
A_grade_percent,B_grade_percent,C_grade_percent,D_grade_percent
```

```python
#8

gender_counts = df.iloc[2:-1,3].value_counts()

total_students = gender_counts.sum()
male = gender_counts['М'] / total_students * 100
female = gender_counts['Ж'] / total_students * 100

print(f"Юношей: {male:.2f}%")
print(f"Девушек: {female:.2f}%")
```
